```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(12, 10))

# Data for the plots
x = np.arange(1, 6)
y = x**2

categories = ['A', 'B', 'C', 'D']
bar_values = [5, 7, 3, 9]

scatter_x = np.array([1, 2, 3, 4, 5])
scatter_y = np.array([2, 4, 6, 8, 10])

pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]

# Subplot 1: Line Chart
plt.subplot(2, 2, 1)
plt.plot(x, y, 'b-o')
plt.title('Line Chart')
plt.xlabel('X')
plt.ylabel('Y = X')
plt.grid(True)

# Subplot 2: Bar Chart
plt.subplot(2, 2, 2)
plt.bar(categories, bar_values, color=['#3498db', '#e74c3c', '#2ecc71', '#f39c12'])
plt.title('Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')

# Subplot 3: Scatter Plot
plt.subplot(2, 2, 3)
plt.scatter(scatter_x, scatter_y, color='red', s=100)
plt.title('Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)

# Subplot 4: Pie Chart
plt.subplot(2, 2, 4)
plt.pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Pie Chart')

plt.tight_layout()
plt.show()
```
