```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = x**2

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', linewidth=2)

plt.title('Square Function: y = x', fontsize=16)
plt.xlabel('x values', fontsize=12)
plt.ylabel('y values (x)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)

plt.xlim(-0.5, 10.5)
plt.ylim(-5, 105)

markers_x = [0, 2, 4, 6, 8, 10]
markers_y = [x**2 for x in markers_x]
plt.scatter(markers_x, markers_y, color='red', zorder=3)

plt.annotate('(8, 64)', xy=(8, 64), xytext=(6, 70),
             arrowprops=dict(facecolor='black', shrink=0.05, width=1.5))

plt.tight_layout()
plt.show()
```
